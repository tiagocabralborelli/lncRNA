```python
from Bio import SeqIO
```

```python
#Extract uncharacterized rma
with open("C.arabica/c.arabica.uncharacterized.fasta","w+") as fastafile:
    for req in SeqIO.parse("C.arabica/refseq/GCF_003713225.1/rna.fna","fasta"):
        if "uncharacterized" in req.description:
            fastafile.write(f">{req.description}\n")
            fastafile.write(f"{str(req.seq)}\n")
```

```python
!blastn -db C.arabica/rna_db/db -query C.arabica/PXD002963/generated/PXD002963_58016.fasta -outfmt 7 -out C.arabica_rna_blast.tsv
```

```python
import pandas as pd
# Import blast results
c_arabica_rna_blast = pd.read_csv(
    "C.arabica_rna_blast.tsv", 
    sep = "\t", 
    skiprows = 6,
    names = ["query_id", "subject_id", "identity", "alignment_length", "mismatches", "gap_opens", "q_start", "q_end", "s_start", "s_end", "evalue", "bit_score"]
    )
```

```python
c_arabica_rna_blast.dropna(inplace = True)
```
